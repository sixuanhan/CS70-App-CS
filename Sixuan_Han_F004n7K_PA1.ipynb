Most arithmetic operations are defined for NumPy arrays element-wise. Also, if you have two NumPy arrays `A` and `B`, you can perform matrix-matrix (or matrix-vector) multiplication using the `@` operator as `A @ B`. You can compute the transpose of a matrix using `A.T`.


````python
#
# you may want to define some common set up code here
A = np.array([[3], [-2]])
B = np.array([[6, 1], [-3, 5]])
C = np.array([[7, 3], [-2, 9]])
D = np.array([[6, -2], [-3, -4], [9, 8]])
#

# TODO: put your answer below (leave as None if undefined)
four_A = 4 * A
print(f'four_A=\n{four_A}')
five_B_minus_two_C = 5 * B - 2 * C
print(f'five_B_minus_two_C=\n{five_B_minus_two_C}')
CA = C @ A
print(f'CA=\n{CA}')
At_Dt = A.T @ D.T
print(f'At_Dt=\n{At_Dt}')
DC = D @ C
print(f'DC=\n{DC}')
CB = C @ B
print(f'CB=\n{CB}')
C_Bt = C @ B.T
print(f'C_Bt=\n{C_Bt}')
````

## Problem 2 (3 points)
Here we ask you to implement your own matrix-matrix multiplication and matrix transpose operations by calculating the value of each matrix element using its linear algebra definitions and compare the results to the ones you have calculated with the built-in NumPy functions:

a) Write your own matrix-matrix (or matrix-vector) multiplication function.

b) Compute the transpose of a matrix without using .T

````python
def Matrix_Multiplication(M_A, M_B):
    # Check the sizes of M_A and M_B
    if M_A.shape[1] != M_B.shape[0]:
        raise ValueError("Matrix dimensions are not compatible for multiplication")

    # Initialize the result matrix M_AB with appropriate dimensions
    num_rows_A, num_cols_A = M_A.shape
    num_rows_B, num_cols_B = M_B.shape
    M_AB = np.zeros((num_rows_A, num_cols_B))

    # Perform matrix multiplication
    for i in range(num_rows_A):
        for j in range(num_cols_B):
            for k in range(num_cols_A):
                M_AB[i, j] += M_A[i, k] * M_B[k, j]

    return M_AB
````

````python
def Transpose(M):
    # Get the dimensions of the input matrix M
    num_rows, num_cols = M.shape
    
    # Initialize the transpose matrix M_t with switched dimensions
    M_t = np.zeros((num_cols, num_rows), dtype=M.dtype)
    
    # Populate M_t with transposed values
    for i in range(num_rows):
        for j in range(num_cols):
            M_t[j, i] = M[i, j]
    
    return M_t
````

````python
# Now compare the results from your functions to your earlier results from NumPy
CA = Matrix_Multiplication(C, A)
print(f'CA=\n{CA}')
At_Dt = Matrix_Multiplication(Transpose(A), Transpose(D))
print(f'At_Dt=\n{At_Dt}')
DC = Matrix_Multiplication(D, C)
print(f'DC=\n{DC}')
CB = Matrix_Multiplication(C, B)
print(f'CB=\n{CB}')
C_Bt = Matrix_Multiplication(C, Transpose(B))
print(f'C_Bt=\n{C_Bt}')
````

## Problem 3 (3 points)

For the following systems of equations, formulate each as a linear system in python in matrix form as $\mathbf{Mu} = \mathbf{v}$
and solve the linear system (use the NumPy function `np.linalg.solve`). Provide both your code and print your solution to the system of equations.

Solve:

a)
$$
\begin{align}
6x+5y &= 17\\
2x+y &= 5
\end{align}
$$

````python
# put your answer here
M_a = np.array([[6, 5], [2, 1]])
v_a = np.array([[17], [5]])

# Mu = v => u = M^{-1} * v
u_a = np.linalg.inv(M_a) @ v_a

print(f'u_a=\n{u_a}')
````

b)
$$
\begin{align}
-x+3y-3z &= -8\\
4x+5y-z &= 14\\
x+3y+2z &= 16
\end{align}
$$

````python
# put your answer here
M_b = np.array([[-1, 3, -3], [4, 5, -1], [1, 3, 2]])
v_b = np.array([[-8], [14], [16]])
u_b = np.linalg.inv(M_b) @ v_b

print(f'u_b=\n{u_b}')
````

c)
$$
\begin{align}
-4x+2z &= 14\\
-2x-4z &= 22\\
x+y+z &= -12
\end{align}
$$

````python
# put your answer here
M_c = np.array([[-4, 0, 2], [-2, 0, -4], [1, 1, 1]])
v_c = np.array([[14], [22], [-12]])
u_c = np.linalg.inv(M_c) @ v_c

print(f'u_c=\n{u_c}')
````

d)
$$
\begin{align}
-3x+9 &= z\\
3y-6 &= 3x\\
-4x+5y+2z &= 5
\end{align}
$$

````python
# put your answer here
M_d = np.array([[-3, 0, -1], [-3, 3, 0], [-4, 5, 2]])
v_d = np.array([[-9], [6], [5]])
u_d = np.linalg.inv(M_d) @ v_d

print(f'u_d=\n{u_d}')
````
